## 1. 导入模型

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.svm import SVC
```

## 2. 制造数据集

```python
np.random.seed(2)
X = np.r_[np.random.randn(20, 2) - [0, 2], np.random.randn(20, 2) + [0, 2]]
Y = [0] * 20 + [1] * 20
```

我们把数据$X$的第一列，记为$x_1$，第二列记为$x_2$

可视化

```python
def plot_data(X, Y):
    # 新建一个 8 × 8 的图
    plt.figure(figsize = (8, 8))

    # 绘制散点图
    plt.scatter(X[:, 0], X[:, 1], c = Y, cmap = plt.cm.Paired, edgecolors = 'k')

    # 设置横纵坐标标签
    plt.xlabel('x1')
    plt.ylabel('x2')

    # 设定图的范围
    plt.xlim((-4, 4))
    plt.ylim((-4, 4))
```

```python
plot_data(X, Y)
```

## 3. 训练模型

这里我们使用线性核函数，C设定为100

```python
# 创建模型
clf = SVC(kernel = 'linear', C = 100, random_state = 32)

# 训练模型
clf.fit(X, Y)
```

## 4. 绘制超平面

在二维平面中，我们将SVM参数写为$w_0$和$w_1$，截距为$b$。

超平面方程为：

$$
w^{\mathrm{T}} x + b = 0
$$

写成分量形式为：

$$
w_0 x_1 + w_1 x_2 + b = 0
$$

等价于：

$$
x_2 = - \frac{w_0}{w_1} x_1 - \frac{b}{w_1}
$$

所以我们可以在上面的图中，绘制这个分离超平面，上图中，纵坐标为$x_2$，横坐标为$x_1$，我们可以通过SVC的coef_属性获取$w$，通过intercept_获取$b$。

画图的时候我们可以给定任意两点，如x1等于-5和5的点，计算出这两个点的x2值，然后通过连线的方式，绘制出超平面

```python
def compute_hyperplane(model, x1):
    '''
    计算二维平面上的分离超平面，
    我们通过w0，w1，b以及x1计算出x2，只不过这里的x1是一个ndarray，x2也是一个ndarray
    
    Parameters
    ----------
    model: sklearn中svm的模型
    
    x1: numpy.ndarray，如[-5, 5]，表示超平面上这些点的横坐标
    
    Returns
    ----------
    x2: numpy.ndarray, 对应的纵坐标
    
    '''
    w0 = model.coef_[0][0]
    w1 = model.coef_[0][1]
    b = model.intercept_[0]
    
    # YOUR CODE HERE
    x2 = (-w0 / w1) * x1 - (b / w1)
    
    return x2
```

```python
# 测试样例
x1t = np.array([-5, 5])
x2t = compute_hyperplane(clf, x1t)
print(x2t) # [ 2.05713424 -2.27000147]
```

```python
# 绘制数据
plot_data(X, Y)

# 在横坐标上选两个点
x1 = np.array([-5, 5])

# 计算超平面上这两个点的对应纵坐标
x2 = compute_hyperplane(clf, x1)

# 绘制这两点连成的直线
plt.plot(x1, x2, '-', color = 'red')
```

## 5. 绘制间隔

根据SVM的原理，间隔超平面的方程是：

$$
w^{\mathrm{T}}x + b = \pm 1
$$

我们先讨论右侧为1的情况：

$$
w^{\mathrm{T}}x + b = 1
$$

写成分量形式：

$$
w_0 x_1 + w_1 x_2 + b = 1
$$

根据上面第四节的变换，可以得到：

$$
x2 = \frac{1}{w_1} - \frac{w_0}{w_1} x_1 - \frac{b}{w_1}
$$

同理，当右侧为-1时，可得：

$$
x2 = - \frac{1}{w_1} - \frac{w_0}{w_1} x_1 - \frac{b}{w_1}
$$

**可以发现，间隔超平面的方程就是在分离超平面上增加或减去$\frac{1}{w_1}$**

```python
def compute_margin(model, x1):
    '''
    计算二维平面上的间隔超平面，
    我们通过w0，w1，b以及x1计算出x2，只不过这里的x1是一个ndarray，x2也是一个ndarray
    
    Parameters
    ----------
    model: sklearn中svm的模型
    
    x1: numpy.ndarray，如[-5, 5]，表示超平面上这些点的横坐标
    
    Returns
    ----------
    x2_up: numpy.ndarray, 一条间隔超平面上对应的纵坐标
    
    x2_down: numpy.ndarray, 另一条间隔超平面上对应的纵坐标
    
    '''
    
    # 先调用compute_hyperplane计算超平面的纵坐标
    x2 = compute_hyperplane(model, x1)
    
    w1 = model.coef_[0][1]
    
    margin = 1 / np.sqrt(np.sum(model.coef_ ** 2))
    
    # YOUR CODE HERE
    x2_up = x2 + margin / np.sqrt(1 + model.coef_[0][1] ** 2)
    
    # YOUR CODE HERE
    x2_down = x2 - margin / np.sqrt(1 + model.coef_[0][1] ** 2)
    
    return x2_up, x2_down
```

```python
# 测试样例
x1t = np.array([-5, 5])
x2_upt, x2_downt = compute_margin(clf, x1t)
print(x2_upt)     # [ 2.43100836 -1.89612735]
print(x2_downt)   # [ 1.68326012 -2.64387559]
```

```python
# 绘制数据
plot_data(X, Y)

# 在横坐标上选两个点
x1 = np.array([-5, 5])

# 计算超平面上这两个点的对应纵坐标
x2 = compute_hyperplane(clf, x1)

# 计算间隔超平面上这两个点的对应纵坐标
x2_up, x2_down = compute_margin(clf, x1)

# 绘制分离超平面和间隔超平面
plt.plot(x1, x2, '-', color = 'red')
plt.plot(x1, x2_up, 'k--')
plt.plot(x1, x2_down, 'k--')
```

```python
# 绘制数据
plot_data(X, Y)

# 在横坐标上选两个点
x1 = np.array([-5, 5])

# 计算超平面上这两个点的对应纵坐标
x2 = compute_hyperplane(clf, x1)

# 计算间隔超平面上这两个点的对应纵坐标
x2_up, x2_down = compute_margin(clf, x1)

# 绘制分离超平面和间隔超平面
plt.plot(x1, x2, '-', color = 'red')
plt.plot(x1, x2_up, 'k--')
plt.plot(x1, x2_down, 'k--')

# 标出支持向量
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s = 200, facecolors = 'none', edgecolors = 'red')
```

# 作业2：简述为什么要加入软间隔？C的值对SVM有什么影响？

## 在支持向量机（Support Vector Machine，SVM）中，软间隔（soft margin）是为了允许在训练数据中存在一些噪声或异常点（outliers），从而提高模型的鲁棒性。软间隔允许一些样本不满足严格的分类条件，即允许一些样本落在决策边界附近或甚至在错误的一侧。这对于处理具有噪声或异常值的数据集非常重要，因为在实际应用中，数据集往往不是完美的、没有噪声的。\n软间隔的引入通过引入一个参数C来控制，C是一个正则化参数，用于平衡最小化训练误差和最小化模型复杂度之间的权衡。C的值越大，模型越倾向于对训练数据拟合得更好，即更强调对训练数据的正确分类，但可能会导致对噪声或异常点过于敏感，从而使模型泛化能力下降。相反，C的值越小，模型越容忍训练数据中的噪声或异常值，更倾向于保持一个简单的决策边界，有助于提高模型的泛化能力。\n因此，C的选择在SVM中是一个重要的超参数调优问题。在实际应用中，可以通过交叉验证等方法来选择一个合适的C的值，以获得对于特定问题和数据集的最佳性能。
